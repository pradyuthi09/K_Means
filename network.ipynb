```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ziya07/network-traffic-anomaly-detection-dataset")

print("Path to dataset files:", path)
```

```python

import os
os.listdir(path)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.ensemble import IsolationForest
```

```python
data=pd.read_csv(os.path.join(path, "embedded_system_network_security_dataset.csv"))
data.head()
```

```python
data.info()
```

```python
# if columns is bool type then convert it to int type
for col in data.columns:
    if data[col].dtype == 'bool':
        data[col] = data[col].astype(int)
```

```python
# Feature Engineering
#Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_features=scaler.fit_transform(data)
features=pd.DataFrame(scaled_features, columns=data.columns)
features.head()
```

```python
# parameters for isolation forest
n_estimators=100
contamination=0.04
sample_size=256
```

```python
import pandas as pd

scaled_features_df = pd.DataFrame(
    scaled_features,
    columns=features.columns)
```

```python
from sklearn.ensemble import IsolationForest

# Train the Isolation Forest model
model = IsolationForest(
    n_estimators=n_estimators,
    contamination=contamination,
    max_samples=sample_size,
    random_state=42
)

model.fit(scaled_features)
```

```python
# Predict anomalies
anomaly_labels = model.predict(scaled_features)

scaled_features_df['anomaly'] = anomaly_labels


```

```python
# Add anomaly labels to DataFrame
scaled_features_df['anomaly'].value_counts()

```

```python
# visualize the anomalies using  a scatter plot take two features for visualization
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=scaled_features_df,
    x='packet_size',
    y='inter_arrival_time',
    hue='anomaly',
    palette={1: 'blue', -1: 'red'},
    alpha=0.6
)

plt.title('Isolation Forest Anomaly Detection')
plt.xlabel('Packet Size')
plt.ylabel('Inter Arrival Time')
plt.legend(title='Anomaly (1 = Normal, -1 = Anomaly)')
plt.grid(alpha=0.3)
plt.show()
```
